```python
#Importing New Data into Excel File from SQL Queries: sales_datadump

# Opening the Source Excel File

filename1 = r"C:\Users\adamk\Desktop\End-to-End\data\sales_datadump.xlsx"
wb1 = xl.load_workbook(filename1)
ws1 = wb1.worksheets[0]

# Opening the Destination Excel File

filename_d = r"C:\Users\adamk\Desktop\End-to-End\Executive Reporting.xlsx"
wb_d = xl.load_workbook(filename_d)
wsd1 = wb_d.worksheets[2]

# Calculate the Total Number of Rows and Columns in the Source Excel File

mr1 = ws1.max_row
mc1 = ws1.max_column

# Copying the Cell Values from the Source Excel File

for i in range (1, mr1 + 1):
    for j in range (1, mc1 + 1):
        
        # Reading Cell Value from Source Excel File
        
        c1 = ws1.cell(row = i, column = j)
        
        # Writing the Cell Value to the Destination Excel File
        
        wsd1.cell(row = i, column = j).value = c1.value
        
wb_d.save(str(filename_d))
```

```python
# Refresh Pivot Tables with Newly Imported Data

# Initialize a New Instance of Excel

ExcelApp1 = win32com.client.Dispatch("Excel.Application")
wb_ea1 = ExcelApp1.Workbooks.Open(r"C:\Users\adamk\Desktop\End-to-End\Executive Reporting.xlsx")

# If the Source Excel File has Linked Cells, Bypass Request to Update Dialogue Box

ExcelApp1.AskToUpdateLinks = False

# Count the Number of Worksheets in the Source Excel File

wb_ea1_ct = wb_ea1.Sheets.Count

# Loop through the Worksheets and Update those that Contain Pivot Tables

for i in range(wb_ea1_ct):
    ws_ea1 = wb_ea1.Worksheets[i]
    
    # Bypass Protected Worksheets if Applicable
    #ws_ea1.Unprotect()
    
    ws_ea1_pivot_ct = ws_ea1.PivotTables().Count
    for j in range(1, ws_ea1_pivot_ct+1):
        ws_ea1.PivotTables(j).PivotCache().Refresh()
        
    # Reprotect Worksheets if Applicable
    #ws_ea1.Protect(DrawingObjects=True, Contents=True, Scenarios=True, AllowUsingPivotTables=True)

# End the New Instance of Excel
wb_ea1.Save()
ExcelApp1.Quit()
```

```python
# Create a PowerPoint with Output from Excel File

# Create a New Instance of Excel

ExcelApp2 = win32com.client.Dispatch("Excel.Application")

# Open the Source Excel File

wb_ea2 = ExcelApp2.Workbooks.Open(r"C:\Users\adamk\Desktop\End-to-End\Executive Reporting.xlsx")

# If the Source Excel File has Linked Cells, Bypass Request to Update Dialogue Box

ExcelApp2.AskToUpdateLinks = False

# Create a Dictionary Object to Include the Named Ranges (i.e., charts) from the Source Excel File

RangeDict = {}

# Loop through the Named Ranges from the Source Excel File and Store them in the Newly Created Dictionary

for namedRng in wb_ea2.Names:
    
    # Find the Named Range Index and Name
    
    rngIndex = namedRng.Index
    rngName = namedRng.Name
    
    # Set the Index as the Key and the Name as the Value of the Dictionary
    
    RangeDict[rngIndex] = rngName

# Print the Dictionary to Verify if Any Extraneous Named Ranges are in the Workbook

#print(RangeDict)

# Create a Subset of Ranges to Include Only the Named Ranges of the Charts

subset = ['Range1', 'Range2', 'Range3', 'Range4', 'Range5', 'Range6', 'Range7']

# Create a Secondary Dictionary with Only the Ranges from the `subset`

subdict = {k:v for k, v in RangeDict.items() if v in subset}

# Print the `subset` Dictionary to Verify that Only the Desired Named Ranges are Included

#print(subdict)

# Create a List Based on the Secondary Dictionary Key Value Pairs
keylist = list(subdict)

# Declare a Variable that is One Integer Less than the Key for the First Desired Named Range
# This Will Be Used as a Dependency for a Dynamic Calculation When Calling for the Ranges to Print in the PPT Presentation
calckey = keylist[0]-1

# Create a New Instance of PowerPoint

PPTApp1 = win32com.client.Dispatch("PowerPoint.Application")

# Set the Visibility of the PowerPoint Application to True

PPTApp1.Visible = True

# Create a New Presentation in the Application

pr_pp1 = PPTApp1.Presentations.Add()

# Loop through the Items in the `subdict`, Add a Slide, Copy the Output, and Paste it into the Presentation

for key, value in subdict.items():
    
    # Use the Key as the Index when Creating the Slide; the Index Subtracts the Count of Extraneous Ranges
    # Layout = 12 Selects a Blank Layout
    
    sl_pp = pr_pp1.Slides.Add(Index = key-calckey, Layout = 12)
    
    # Copy the Ranges Usings the Value
    
    ExcelApp2.Range(value).Copy()
    
    # DataType = 10 is an OLEObject; Set the Link Equal to False
    
    sl_pp.Shapes.PasteSpecial(DataType = 10, Link = False)
    
pr_pp1.SaveAs(r"C:\Users\adamk\Desktop\End-to-End\Executive Reporting.pptx")

wb_ea2.Save()
pr_pp1.Save()
ExcelApp2.Quit()
PPTApp1.Quit()
```
